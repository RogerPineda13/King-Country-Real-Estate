```python
import pandas as pd
```

```python
df = pd.read_csv("data/kc_house_data.csv", index_col='id')
```

```python
df
```

```python
df['sqft_basement'] = df['sqft_living']-df['sqft_above']
df['Basement'] = None
```

```python
df['Basement'] = df['sqft_basement'].map(lambda x: False if x == 0 else True)
df['Basement'] = df['Basement'].astype(int)
df
```

```python
df['Basement'].value_counts()
```

```python
df['datetime']=pd.to_datetime(df['date'])
```

```python
df['soldyear'] = df['datetime'].dt.year
```

```python
df['age_when_sold'] =  df['soldyear'] - df['yr_built']
df
```

```python
df['bathrooms'].value_counts()
```

```python
df['bathrooms'] = df['bathrooms'].replace(0.75, 1)
df['bathrooms'] = df['bathrooms'].replace(1.75, 2)
df['bathrooms'] = df['bathrooms'].replace(2.75, 3)
df['bathrooms'] = df['bathrooms'].replace(3.75, 4)
df['bathrooms'] = df['bathrooms'].replace(4.75, 5)
df['bathrooms'] = df['bathrooms'].replace(5.75, 6)
df['bathrooms'] = df['bathrooms'].replace(6.75, 7)
df['bathrooms'] = df['bathrooms'].replace(7.75, 8)
df['bathrooms'] = df['bathrooms'].replace(1.25, 1.5)
df['bathrooms'] = df['bathrooms'].replace(2.25, 2.5)
df['bathrooms'] = df['bathrooms'].replace(3.25, 3.5)
df['bathrooms'] = df['bathrooms'].replace(4.25, 4.5)
df['bathrooms'] = df['bathrooms'].replace(5.25, 5.5)
df['bathrooms'] = df['bathrooms'].replace(6.25, 6.5)
```

```python
df['bathrooms'].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
bathrooms = df[['bathrooms']]
ohe = OneHotEncoder(categories='auto', sparse=False, handle_unknown='ignore')
bathrooms_transform = ohe.fit_transform(bathrooms)
bathroomsdf = pd.DataFrame(bathrooms_transform, columns=ohe.categories_[0], index=bathrooms.index)
```

```python
df.drop(columns='bathrooms', inplace=True)
```

```python
df = pd.concat([df, bathroomsdf], axis=1)
df
```
